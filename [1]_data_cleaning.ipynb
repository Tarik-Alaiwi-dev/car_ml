```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv(r'data/Ad_table.csv')
df.head()
```

```python
df.tail()
```

## Data type standarization

```python
df['Runned_Miles'] = df['Runned_Miles'].str.replace('1 mile', '1').astype(float)
```

```python
df['Engine_size'] = df['Engine_size'].str.replace('L', '').astype(float)
```

## Standarize column names

```python
df['Engine_size'] = df['Engin_size']
```

```python
df['Price'] = df['Price']
```

```python
df['Prod_year'] = df['Reg_year']
```

```python
df['Model'] = df['Genmodel']
```

```python
df['Body_type'] = df['Bodytype']
```

```python
df['Runned_miles'] = df['Runned_Miles']
```

```python
df.drop(columns=['Engin_size', 'Genmodel_ID', 'Adv_ID', 'Reg_year', 'Genmodel', 'Bodytype', 'Runned_Miles'], inplace=True)
```

## Handle missing values

```python
df = df[df['Price'] != 'Uknown']
df['Price'] = df['Price'].astype(int)
```

```python
df.head()
```

## Handle duplicates

```python
df.drop_duplicates(subset=['Maker', 'Color', 'Runned_miles', 'Gearbox', 
                           'Fuel_type', 'Price', 'Seat_num', 'Door_num', 
                           'Engine_size', 'Prod_year', 'Model', 'Body_type'], keep=False, inplace=True)
```
